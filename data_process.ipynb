```python
from ultralytics import YOLO
import os
import cv2
from pathlib import Path
import uuid
```

```python
data_path = Path("data")
for video_name in os.listdir(data_path):
    folder_name = video_name.split(".")[0]
    if folder_name == "demo":
        continue
    video_path = data_path / video_name
    model = YOLO("weights/vehicle_detection.pt")
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    skip = 0.2
    skip_frame = max(1, int(fps*skip))
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_count += 1
        if frame_count % skip_frame != 0:
            continue
        result = model.track(frame, conf=0.75, iou=0.7, tracker="weights/tracker/botsort.yaml", persist=True, device="cpu", verbose=False)[0]
        for bbox in result.boxes:
            if bbox.id is not None:
                x1, y1, x2, y2 = bbox.xyxy[0].cpu().numpy().astype(int)
                obj = frame[y1:y2, x1:x2]
                id_ = bbox.id[0].cpu().numpy().astype(int)
                folder_id_path = f"LicensePlate/data/temp/{folder_name}/{folder_name}_{id_}"
                os.makedirs(folder_id_path, exist_ok=True)
                random_id_img = str(uuid.uuid4())
                cv2.imwrite(f"{folder_id_path}/{random_id_img}.jpg", obj)
    cap.release()
    print(f"Finished extract object from {video_name}!!!")     
```
